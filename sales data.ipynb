```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
```

```python
d=pd.read_csv('SalesJan2009.csv',quotechar='"')
```

```python
d.columns
```

```python
d.Payment_Type.value_counts()
a=d.Payment_Type.value_counts().sort_values(ascending=True)
a.plot.barh(stacked=True, alpha=0.5)
plt.title('Payment Methods')
plt.show()
```

```python
d.Name.value_counts()
g=d.groupby(['Name','Product'])
```

```python
functions=['count']
result=g['Product'].agg(functions)
# result=d['Price','Product'].agg(functions)
#g=d.groupby(['Price','Product'])
result.head(20).plot.barh()
plt.title('Person Name \n with \n The Product Quantity')
plt.show()
```

```python
h=d.groupby(['Product','Name','City'])
```

```python
results=h['Product'].agg(functions)
```

```python
results
```

```python
results.describe().plot.barh()
plt.show()
```

```python
def dat_slicing(x):
    if x:
        x=x[0:7]
        return x
```

```python
d['Transaction_date']=d['Transaction_date'].map(dat_slicing)
```

```python
p_d=d.groupby(['Product','Transaction_date'])
res=p_d['Product'].agg(functions)
res.head(20).plot.barh()
plt.title('Product sales By Dates')
plt.show()
```

```python
res.describe().plot.barh()
plt.show()
```

```python
f_g=d.groupby('Product')['Price'].describe()
```

```python
f_g.plot.barh()
plt.title('Product Sold')
plt.show()
```
